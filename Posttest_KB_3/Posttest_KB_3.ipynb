### Melihat kan data Restaurant

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Restaurant reviews.csv')
df_Restaurant = pd.read_csv('Restaurant reviews.csv')
df_Restaurant.head()
```

### Mengelompokkan data berdasarkan Rating

```python
data = df_Restaurant.groupby('Rating').mean()
data
```

### Gambar Grafik Batang dalam Restaurant

```python
x = data.index
y = data.Pictures
plt.bar(x, y, color=['Red', 'Green', 'Yellow'])
plt.title('Rata-rata panjang Like')
plt.xlabel('Banyak Like')
plt.ylabel('panjang rata-rata Like')
plt.show()
```

### Gambar Grafik Histogram

```python
plt.hist(df_Restaurant['Pictures'], bins=10, color='Yellow')
plt.title('Histogram panjang petal')
plt.ylabel('Frekuensi kemunculan')
plt.xlabel('Panjang petal')
plt.show()
```

### Gambar Pie Chart dalam Restaurant

```python
data = df.Rating.value_counts()
size = 350
plt.pie(data, labels=data.index, autopct="%1.1f%%", startangle=140, )
plt.title('Jumlah Data')
plt.show()
```
