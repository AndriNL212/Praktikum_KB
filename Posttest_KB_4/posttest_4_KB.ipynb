# Mengatasi nilai null pada kolom Backers yang bertipe bilangan bulat (integer)# 
Metode Median

```python
dataset2.loc[200000:, 'Pictures'] = np.nan

print("Jumlah Nilai Null pada kolom  Backers sebelum >>", dataset2['Pictures'].isna().sum())

dataset2['Pictures'].fillna(dataset2['Pictures'].median(), inplace=True)

print("\nJumlah Nilai Null pada kolom  Pictures Setelah >>", dataset2['Pictures'].isna().sum())
```

# Metode Modus

```python
dataset2.loc[200000:, 'Pictures'] = np.nan

print("Jumlah Nilai Null pada kolom  Pictures sebelum >>", dataset2['Pictures'].isna().sum())

dataset2['Pictures'].fillna(dataset2['Pictures'].mode()[0], inplace=True)

print("\nJumlah Nilai Null pada kolom  Pictures setelah >>", dataset2['Pictures'].isna().sum())
```

# Metode Mean

```python
dataset2.loc[200000:, 'Pictures'] = np.nan

print("Jumlah Nilai Null pada kolom  Pictures sebelum >>", dataset2['Pictures'].isna().sum())

dataset2['Pictures'].fillna(dataset2['Pictures'].mean(), inplace=True)

print("\nJumlah Nilai Null pada kolom  Pictures setelah >>", dataset2['Pictures'].isna().sum())
```
