### Mendeklarasikan jumlah record, jumlah attribute, list nama atribut, dan juga jumlah attribute angka ke dalam sebuah variable, kemudian memasukkannya ke dalam sebuah dictionary dan menampilkan

```python
jumlah_seluruh_record = len(dataset)
jumlah_seluruh_attribute = len(dataset.columns)
nama_attribute = list(dataset.columns)
attribute_angka = len( dataset.select_dtypes(include=['number']).columns)
dictionary = {
    'Jumlah record' : jumlah_seluruh_record,
    'Jumlah seluruh attribute' : jumlah_seluruh_attribute,
    'nama - nama attribute' : nama_attribute,
    'Jumlah attribute dengan tipe angka' : attribute_angka
}
for key, value in dictionary.items():
    print(f'{key} : {value}')
```

### menampilkan jumlah seluruh nilai dari semua attribut yang bernilai object

```python
attribute_object = dataset.select_dtypes(include=['object']).columns.tolist()
for i in attribute_object :
    print(dataset[i].value_counts())
```
